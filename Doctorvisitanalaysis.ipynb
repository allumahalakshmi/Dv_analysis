```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DoctorVisits - DA.csv")
print(df.head(15))
```

```python
df.info()
df["illness"].value_counts()
df["gender"].value_counts()
```

```python
y=list(df.income)
plt.boxplot(y)
plt.show()
```

```python
df.groupby(['gender','reduced']).mean()
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')

plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')

```

```python
sns.histplot(df.gender,bins=2)
```

```python
label=['yes','no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
x =[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health Insurance due to low income")
plt.show()
```

```python
# % of people having private Insurance
Y = df[df['private']=='yes']
N = df[df['private']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people having private Insurance ")
plt.show()
```

```python
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt Insurance due to old age,disability or veteran status ")
plt.show()
```

```python
#plot a horizontal bar chart to analyse the reduced days of activity due to illness based on Gender
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
```

```python
#creating the bar chart
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
```

```python
plt.title('Bar Chart')
plt.xlabel('Reduced activity')
plt.ylabel('gender')
#show the plot
plt.show()

```
